### Stretching and shrinking

```python
from PIL import Image
image = Image.open('lenna.png')

new_width , new_height = 512,200

streached_image = image.resize((new_width , new_height))

streached_image.save('streached_image.jpg')

streached_image
```

```python
import cv2

image = cv2.imread('lenna.png')

# Define new dimensions(smaller than original)
new_width , new_height = 512,200
# Stretch image using interpolation
streached_image = cv2.resize(image,dsize = (new_width , new_height),interpolation=cv2.INTER_CUBIC)

cv2.imshow('Stretched Image',streached_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread('lenna.png')

# Define new dimensions(larger than original)
new_width , new_height = 1700,1000

# Stretch image using interpolation
streached_image = cv2.resize(image,dsize = (new_width , new_height),interpolation=cv2.INTER_CUBIC)

cv2.imshow('Stretched Image',streached_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Histogram Equalization

```python
import numpy as np

img = cv2.imread('lenna.png',0)

equ = cv2.equalizeHist(img)

res = np.hstack((img,equ))

cv2.imshow('image',res)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
res = np.vstack((img,equ))

cv2.imshow('image',res)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
plt.title('Original Histogram')
plt.xticks(range(0,256,10))
plt.hist(img.flatten(),bins=range(0,256));
plt.subplot(1,2,2)
plt.title('Equilized Histogram')
plt.xticks(range(0,256,10))
plt.hist(equ.flatten(),bins=range(0,256));
```
