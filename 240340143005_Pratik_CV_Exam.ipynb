## Rotate the image by 45 degrees

```python
# Rotate the image
(h, w) = img.shape[:2]
center = (w / 2, h / 2)
M = cv2.getRotationMatrix2D(center, 45, 1)
rotated = cv2.warpAffine(img, M, (w, h))

# Display the images
cv2.imshow('Rotated', rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Scale the image by a factor of 1.5

```python
scaled = cv2.resize(rotated, None, fx=1.5, fy=1.5)

cv2.imshow('Scaled', scaled)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Display the original image and the trasnformed image side by side

```python
cv2.imshow('Original', img)
cv2.imshow('Transformed', scaled)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Read an image from the file system

### Image reading by matplotlib

```python
image = plt.imread('Q2_CV_Mar2014.jpg')
```

```python
plt.imshow(image)
```

### Image reading by cv2

```python
image1 = cv2.imread('Q2_CV_Mar2014.jpg')

cv2.imshow('Image',image1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Convert the image to the HSV color space

```python
kernal = np.ones((3,3),np.uint8)

# converting the image to HSV format
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Displaying the image
cv2.imshow('My Image',image)
cv2.imshow('HSV Image',hsv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Create a mask to extract a specific color range from the image with Lower HSV value: [35, 100, 100] and Upper HSV value: [85, 255, 255]

```python
# defining the lower and upper values of HSV
Lower_hsv = np.array([35, 100, 100])
Upper_hsv = np.array([85, 255, 255])

# creating the mask
Mask = cv2.inRange(hsv, Lower_hsv, Upper_hsv)

# Inverting the mask by performing bitwise-not opoeration
Mask = cv2.bitwise_not(Mask)

# Displaying the image
cv2.imshow('My Image',hsv)
cv2.imshow('Mask', Mask)
cv2.waitKey(0)

# closing all open windows
cv2.destroyAllWindows()
```

## Display the original image and the mask image side by side

```python
# Displaying the image
cv2.imshow('Original Image',image)
cv2.imshow('HSV image', hsv)
cv2.imshow('Mask', Mask)
cv2.waitKey(0)

# closing all open windows
cv2.destroyAllWindows()
```
